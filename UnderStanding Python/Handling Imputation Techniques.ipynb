# Mean method

```python
# whenever you have column values which is evenly distributed then you can use mean method to fill null values.

# for exaple if you have the data which is distributed within the range of 1-10 then you can use mean method to fill nul values.
```

## median method

```python
# whenever you have varying out of range, then you can use median method to fill null values.

# example , in the data if some values are in range of 1-100 and some are out of range 10000 then we should have to use median techniques.
```
